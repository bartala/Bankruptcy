```python
# get company ticker (stock symbol)
def getTicker(company_name):
    yfinance = "https://query2.finance.yahoo.com/v1/finance/search"
    user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'
    params = {"q": company_name, "quotes_count": 1, "country": "United States"}

    res = requests.get(url=yfinance, params=params, headers={'User-Agent': user_agent})
    data = res.json()

    company_code = data['quotes'][0]['symbol']
    return company_code
```

```python
tickers = []

for comp in company_name:
  try:
    x = getTicker(comp)
    print(x)
  except Exception as e:
    x = ''
    print(e)
  tickers.append(x)
```

```python
# We can get data by our choice by giving days bracket
start_date= "2020-01-01"
end_date="2021-01-01"

files=[]

# Create a data folder in your current dir.
def SaveData(df, filename):
 df.to_csv(os.path.join(PTH,filename+'.csv'), index = None)


def getData(ticker, files):
  print(ticker)
  data = pdr.get_data_yahoo(ticker, start=start_date, end=end_date)
  data['date'] = (data).index
  data.index.name = None
  files.append(ticker)
  SaveData(data, ticker)


for tik in ticker_list:
 getData(tik, files)
```

```python
df1 = pd.read_csv(os.path.join(PTH,str(files[0])+'.csv'))
df1['ticker'] = files[0]

for i in range(1,len(ticker_list)):
 tmp= pd.read_csv(os.path.join(PTH,str(files[i])+'.csv'))
 tmp['ticker'] = files[i]
 df1 = pd.concat([df1, tmp]).reset_index(drop=True)
```
